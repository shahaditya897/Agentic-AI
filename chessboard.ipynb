```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

rows = int(input("Rows: "))
columns = int(input("Columns: "))

first = input("Color 1: ")
second = input("Color 2: ")

image = np.zeros((rows, columns))

for i in range(rows):
    for j in range(columns):
        value = (i + j) % 2
        image[i, j] = value

plt.imshow(image, cmap=ListedColormap([first, second]))
plt.xticks([])
plt.yticks([])
plt.show()
```
