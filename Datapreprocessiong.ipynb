```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Create Sample Dataset
student = {
    "Name": ["Riya", "Arjun", np.nan, "Meera", "Vivek", "Pooja", "Arjun", "Kunal"],
    "Age": [20, 21, 22, np.nan, 23, 21, 21, np.nan],
    "Gender": ["Female", "Male", "Female", "Female", np.nan, "Female", "Male", "Male"],
    "Marks": [78, 85, np.nan, 90, 88, 76, 85, 82],
    "City": ["Delhi", "Mumbai", "Pune", "Delhi", np.nan, "Pune", "Mumbai", "Delhi"]
}

df = pd.DataFrame(student)

print("Original Data")
print(df)

# Check Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Fill Missing Values
df["Name"] = df["Name"].fillna("Student")
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
df["City"] = df["City"].fillna(df["City"].mode()[0])

print("\nAfter Filling Missing Values")
print(df)

# Remove Duplicate Rows
df = df.drop_duplicates()

print("\nAfter Removing Duplicates")
print(df)

# Check Data Types
print("\nData Types")
print(df.dtypes)

# Convert Age to Integer
df["Age"] = df["Age"].astype(int)

print("\nAfter Datatype Conversion")
print(df.dtypes)

# Encode Gender
df["Gender"] = df["Gender"].map({
    "Male": 1,
    "Female": 0
})

print("\nAfter Encoding Gender")
print(df)

# Normalize Marks
df["Marks"] = (df["Marks"] - df["Marks"].min()) / (df["Marks"].max() - df["Marks"].min())

print("\nAfter Normalization")
print(df)

# Features and Target
X = df[["Age", "Gender", "Marks"]]
y = df["City"]

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=1
)

print("\nTraining Data :", len(X_train))
print("Testing Data :", len(X_test))

# Final Dataset
print("\nFinal Preprocessed Dataset")
print(df)
```
