Player Score and Performance

```python
# Number of Players
n = int(input("Enter number of players: "))

for i in range(n):
    print("\nPlayer", i + 1)

    match1 = int(input("Enter Match 1 Score: "))
    match2 = int(input("Enter Match 2 Score: "))
    match3 = int(input("Enter Match 3 Score: "))

    average = (match1 + match2 + match3) / 3

    if average >= 90:
        performance = "Excellent"
    elif average >= 70:
        performance = "Good"
    else:
        performance = "Needs Improvement"

    print("Average Score =", average)
    print("Performance =", performance)
```

Sum of Even Numbers Only

```python
numbers = [12, 7, 20, "AI", 15, 18, "", 10]

total = 0

for i in numbers:
    if type(i) == int and i % 2 == 0:
        total += i

print("Sum =", total)
```

Common Subject Names

```python
subjects1 = ["Maths", "Science", "English", "Computer", "History"]
subjects2 = ["Science", "Gujarati", "Computer", "Maths", "Hindi"]

common = []

for subject in subjects1:
    if subject in subjects2:
        common.append(subject)

print("Common Subjects =", common)
```
