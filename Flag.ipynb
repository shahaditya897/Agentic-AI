```python
import numpy as np
import matplotlib.pyplot as plt

# User Input
height = int(input("Enter Height: "))
width = int(input("Enter Width: "))

# Create white image
flag = np.ones((height, width, 3), dtype=np.uint8) * 255

# Saffron
flag[:height//3] = [255, 153, 51]

# Green
flag[2*height//3:] = [19, 136, 8]

# Ashoka Chakra
cx = width // 2
cy = height // 2
radius = min(height, width) // 10

# Blue Circle
circle = plt.Circle((cx, cy), radius, fill=False,
                    color='navy', linewidth=2)

# Draw 24 spokes
fig, ax = plt.subplots(figsize=(8,5))
ax.imshow(flag)

for i in range(24):
    angle = 2 * np.pi * i / 24
    x = cx + radius * np.cos(angle)
    y = cy + radius * np.sin(angle)
    ax.plot([cx, x], [cy, y], color='navy', linewidth=1)

ax.add_patch(circle)

ax.set_xticks([])
ax.set_yticks([])
ax.set_title("Indian Flag")

plt.show()
```
